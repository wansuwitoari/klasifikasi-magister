```python
df["Bidang_Minat"] = df["Bidang_Minat"].astype(str).astype(float)
df["Motivasi_Studi"] = df["Motivasi_Studi"].astype(str).astype(float)
df["Motivasi_Beasiswa"] = df["Motivasi_Beasiswa"].astype(str).astype(float)
df["Pengalaman_Penelitian"] = df["Pengalaman_Penelitian"].astype(str).astype(float)
df["Komunikasi"] = df["Komunikasi"].astype(str).astype(float)
df["Problem_Solving"] = df["Problem_Solving"].astype(str).astype(float)
df["Literatur_Review"] = df["Literatur_Review"].astype(str).astype(float)
df["Team_Work"] = df["Team_Work"].astype(str).astype(float)
```

```python
X = df.drop("Kategori_Lulus",axis =1)
y = df["Kategori_Lulus"]

print(y)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

acc = accuracy_score(y_test,y_pred)
print(acc)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import cross_val_score

cross_val_score(model, X, y, cv=5)
```

```python
cross_val_score(model, X, y, cv=5).mean()
```
